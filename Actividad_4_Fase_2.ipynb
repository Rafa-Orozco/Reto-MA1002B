```python
import math  

def func_x( x, y): 
    return (x*(1-(x/500))-0.02*x*y) 
def func_y( x, y ): 
    return (-y+0.01*x*y) 

def euler( t0, y0, x0, h, tf ): 
    valuesx = []
    valuesy = []
    valuest = []
    while t0 < tf: 
        y0 = y0 + h * func_y(x0, y0) 
        x0 = x0 + h * func_x(x0, y0) 
        t0 = t0 + h 
        valuesx.append(x0)
        valuesy.append(y0)
        valuest.append(t0)
    return (valuesx,valuesy,valuest)
```

```python
x,y,t= euler(0, 20, 20, 0.01, 30)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import integrate
import ipywidgets as ipw

alpha = 1. #mortality rate due to predators
beta = 0.02
delta = 1.
gamma = 0.01
x0 = 20.
y0 = 20.

def derivative(X, t, alpha, beta, delta, gamma):
    x, y = X
    dotx = (x*(1-(x/500))-0.02*x*y) 
    doty = (-y+0.01*x*y) 
    return np.array([dotx, doty])


Nt = 1000
tmax = 30.
t = np.linspace(0.,tmax, Nt)
X0 = [x0, y0]
res = integrate.odeint(derivative, X0, t, args = (alpha, beta, delta, gamma))
x, y = res.T


plt.figure()
plt.grid()
plt.title("odeint method")
plt.plot(t, x, '-b', label = 'Presas')
plt.plot(t, y, '-r', label = "Prededadores")
plt.xlabel('Time t, [days]')
plt.ylabel('Population')
plt.legend()

plt.show()
```
